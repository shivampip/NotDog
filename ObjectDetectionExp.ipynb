```
def show(img):
    ip.plot_np_imgs(np.expand_dims(img, 0), count= 1)
```

```
def draw_rec(image, x, y, w, h, t=5, col= [0,255,0]):
    img= np.array(image)
    img[x: x+h, y:y+t]= col
    img[x: x+t, y:y+w]= col
    img[x: x+h+t, y+w: y+w+t]= col
    img[x+h: x+h+t , y: y+w+t]= col
    show(img)
    #return img
```

```
def crop(image, thr= 10):
    img= np.array(image)
    xx,yy,_= img.shape
    img= 255- img
    x,y,x2,y2=0,0,0,0
    
    for i in range(xx):
        yog= sum(img[i,:,0])
        if(yog>thr):
            x= i
            break
    for i in range(xx-1, 0, -1):
        yog= sum(img[i,:,0])
        if(yog>thr):
            x2= i
            break
    for i in range(yy):
        yog= sum(img[:,i,0])
        if(yog>thr):
            y= i
            break
    for i in range(yy-1, 0, -1):
        yog= sum(img[:,i,0])
        if(yog>thr):
            y2= i
            break
            
    img= 255-img 
    draw_rec(img, x, y, y2-y, x2-x)        
```
